#### 1.1 
##### By definition, $\|x\|_{2} = \sqrt{\sum\limits_i{|x_i|^2}}$, $\|x\|_{\infty} = \max\limits_i(|x_i|)$. Then we have to find $C_1(n) > 0$, $C_2(n) > 0$  such that for any $x \in R^n$: 
$C_1 \sqrt{\sum\limits_i{|x_i|^2}} \leq \max\limits_i(|x_i|) \leq C_2 \sqrt{\sum\limits_i{|x_i|^2}}$.
##### From $\sqrt{\sum\limits_i{|x_i|^2}} \leq \sqrt{n \cdot (\max\limits_i{|x_i|})^2} = \sqrt{n} \cdot \max\limits_i|x_i|, ($since $ \max\limits_i|x_i| > 0 $ and is a real number$), C_1 = \dfrac{1}{\sqrt{n}}$.
##### And since $\max\limits_i(|x_i|) \leq \sqrt{\sum\limits_i{|x_i|^2}}$ (and there can be equality if only one element of a vector is non-zero), then $C_2 = 1$.
#### Answer: $C_1 = \dfrac{1}{\sqrt{n}}, C_2 = 1$
#### 1.2
##### We will use these facts: 
##### 1. $\| A \|_F^2 = trace(A^*A)$
##### 2. $trace(BC) = trace(CB)$ for any matrices $B$, $C$ that can be multiplied in both ways.
##### Then:
##### $\| U A \|_F^2 = trace((UA)^*\ UA) = trace(A^*U^*UA) = trace(A^*IA) = trace(A^*A) = \| A \|_F^2 \Rightarrow \| U A \|_F = \| A \|_F$
#### $\| A U \|_F^2 = trace((AU)^*\ AU) = trace(AU(AU)^*) = trace(AUU^*A^*) = trace(AA^*) = trace(A^*A) = \| A \|_F^2 \Rightarrow \| A U \|_F = \| A \|_F$
#### 1.3
##### For a vector: l2-norm can be computed as: $ \|x \|_2 = \sqrt{(x, x)}$, and a scalar product as:  $(x, y) = x^*y$. Using this:
$\| U A \|_2 = \sup\limits_{x \neq 0}\dfrac{\|UAx \|_2}{\|x \|_2} = \sup\limits_{\|x \|_2=1}\dfrac{\|UAx \|_2}{\|x \|_2} = \sup\limits_{\|x \|_2=1}{\|UAx \|_2} = \sup\limits_{\|x \|_2=1}{\sqrt{(UAx, UAx)}} = \sup\limits_{\|x \|_2=1}{\sqrt{x^*A^*U^*UAx}} = \sup\limits_{\|x \|_2=1}{\sqrt{x^*A^*Ax}} = \sup\limits_{\|x \|_2=1}{\sqrt{(Ax)^*Ax}} = \sup\limits_{\|x \|_2=1}{\|Ax \|_2} = \sup\limits_{x \neq 0}\dfrac{\|Ax \|_2}{\|x \|_2} = \|A \|_2$ 
#### Going to the next part of equality:
$\| A U \|_2 = \sup\limits_{x \neq 0}\dfrac{\|AUx \|_2}{\|x \|_2} = (*)$.
##### Let $y = Ux$, then $U^*y = U^*Ux$, and $x = U^*y$. Using unitary invariance of l2-norm ($U^*y = y$):
$(*) = \sup\limits_{y \neq 0}\dfrac{\|Ay \|_2}{\|U^*y \|_2} = \sup\limits_{y \neq 0}\dfrac{\|Ay \|_2}{\|y \|_2} = \| A \|_2$.

##### $\| A \|_F = \| U \Sigma V^*\|_F = \|\Sigma\|_F = \sqrt{trace(\Sigma^* \Sigma)} = \sqrt{|\sigma_1|^2 + |\sigma_2|^2 + ... + |\sigma_r|^2} \leq \sqrt{r \max\limits_{i \in {1,2...r}}{|\sigma_i|}^2} = \sqrt{r} \cdot |\sigma_1|$, where $r = rank(A)$ and since we know $\sigma_1 \geq \sigma_2 \geq ... \sigma_r$.
##### $\| A \|_2 = \sup\limits_{x \neq 0}\dfrac{\|Ax \|_2}{\|x \|_2} = \sup\limits_{\| x \|_2=1}{\|Ax \|_2} = \sup\limits_{\| x \|_2=1}{\|U \Sigma V^*x \|_2} = \sup\limits_{\| x \|_2=1}{\|\Sigma x \|_2} = \sqrt{z^*z} = \sqrt{\max\limits_{i \in {1, 2,...r}}|\sigma_i|^2} = \max\limits_{i \in {1, 2,...r}}{|\sigma_i|} = |\sigma_1|$, where $z$ is a vector with the only non-zero element equal to maximal $\sigma_i$ element of diagonal matrix $\Sigma$.

$\| x \|_2 = \sqrt{\sum\limits_{i=1}^{n} |x_i|^2}$
##### and the fact that $\| A \|_2 = \sigma_{max} (A) = \sigma_1 (A)$

##### Therefore, we got:
$\| AB \|_F \leq \|A\|_2 \| B \|_F$

_1._ (10 pts) Consider the following function

$$ F(U, V) = \frac{1}{2}\|X - UV\|_F^2, $$

#### 1
$ F(U, V) = \dfrac{1}{2}\|X - UV\|_F^2 = \dfrac{1}{2} \Big({\sum\limits_{i=1}^{n}{\sum\limits_{j=1}^{n}{(x_{ij} - \sum\limits_{t=1}^{k}u_{it}v_{tj})^2}}}\Big)$

##### Let's use the following facts:
#### $trace(AB) = trace(BA)$, if both $AB$ and $BA$ are defined. 

$\dfrac{d(trace(AB))}{dA} = B^T$

$\dfrac{d(trace(A^T B)}{dA} = B$

#### Moreover, if $B$ is a square matrix,
$\dfrac{d(trace(A^T A B)}{dA} = AB + AB^T$

#### This can be easily proven, for example, let's prove the first formula:
$(AB)_{ii} = \sum\limits_{k=1}^{n} a_{ik} b_{ki}$

$trace(AB) = \sum\limits_{i=1}^{n} \sum\limits_{k=1}^{n} a_{ik} b_{ki}$

$\dfrac{d(trace(AB))}{da_{lm}} = b_{ml} \Rightarrow \dfrac{d(trace(AB))}{dA} = B^T$

##### First,

$\dfrac{dF(U, V)}{dU} = \dfrac{\dfrac{1}{2}d(\|X - UV\|_F^2)}{dU} = \dfrac{1}{2}\dfrac{d(trace(X - UV)^*(X - UV))}{dU} = \dfrac{1}{2}\dfrac{d(trace(X - UV)^T(X - UV))}{dU} = \dfrac{1}{2}\dfrac{d(trace(X^T - (UV)^T)(X - UV))}{dU} = \dfrac{1}{2}\dfrac{d(trace(X^T X) - trace(X^T UV) - trace((V^T U^T X) + trace(V^T U^T UV)}{dU} = \dfrac{1}{2} (0 - \dfrac{d(trace(UV X^T))}{dU} - \dfrac{d(trace((X V^T U^T))}{dU} + \dfrac{d(trace(V^T U^T UV))}{dU}) = \dfrac{1}{2} (-(V X^T)^T - XV^T + UV V^T + U(VV^T)^T) = (UV - X) V^T$



$\dfrac{dF(U, V)}{dU} = (UV - X) V^T$

##### Second,

$\dfrac{dF(U, V)}{dV} = \dfrac{\dfrac{1}{2}d(\|X - UV\|_F^2)}{dV} = \dfrac{1}{2}\dfrac{d(trace(X^T - (UV)^T)(X - UV))}{dV} = \dfrac{1}{2}\dfrac{d(trace(X^T X) - trace(X^T UV) - trace((V^T U^T X) + trace(V^T U^T UV)}{dV} = \dfrac{1}{2} (0 - (X^T U)^T -  U^T X + U^T UV + (U^T U)^TV) = U^T (UV - X)$



$\dfrac{dF(U, V)}{dV} = U^T (UV - X)$

##### Talking about computational complexity, $UV$ takes $O(n^2 k)$, since resulting matrix is $n \times n$ and each element requires $k$ multiplications and $k-1$ additions, $UV - X$ requires $O(n^2)$ and  $(UV - X) V^T$, as well as $U^T (UV - X)$, in total, requires $O(n^2 k)$ (multiplication of matrices $k \times n$ and $n \times n$).

````python
import matplotlib.pyplot as plt
%matplotlib inline
import jax
import jax.numpy as jnp
import time
import numpy as np

import warnings
warnings.filterwarnings("ignore")

from jax.config import config
config.update("jax_enable_x64", True)
````

````python
@jax.jit # Just-in-time compilation
def f(X, U, V):
    matr = X - U @ V
    res = (jnp.linalg.norm(matr, ord='fro'))**2 / 2
    return res

gradu = jax.grad(f, argnums=1, has_aux=False)
gradv = jax.grad(f, argnums=2, has_aux=False)
````

````python
size_range = [200, 500, 1000, 1500, 2000, 3000]
time_analyt = np.zeros((len(size_range), 2))
time_comput = np.zeros((len(size_range), 2))
time_jit = np.zeros((len(size_range), 2))
````

````python
import timeit
for i in range(len(size_range)):
    n = size_range[i]
    k = 100
    X = jax.random.normal(jax.random.PRNGKey(0), (n, n))
    U = jax.random.normal(jax.random.PRNGKey(0), (n, k))
    V = jax.random.normal(jax.random.PRNGKey(0), (k, n))
    #analyt_grad_u = (U @ V - X) @ V.T
    #analyt_grad_v = U.T @ (U @ V - X) 

    t1 = %timeit -o (U @ V - X) @ V.T
    t2 = %timeit -o U.T @ (U @ V - X)
    time_analyt[i, 0] = t1.average
    time_analyt[i, 1] = t2.average
    #time_analyt[i] += time.time() - start_time
    #print(t, i, time.time() - start_time)

    
    t3 = %timeit -o gradu(X, U, V).block_until_ready()
    t4 = %timeit -o gradv(X, U, V).block_until_ready()
    time_comput[i, 0] = t3.average
    time_comput[i, 1] = t4.average
    #print(t, i, time.time() - start_time)
    
    jit_gradu = jax.jit(gradu)
    jit_gradv = jax.jit(gradv)

    t5 = %timeit -o jit_gradu(X, U, V).block_until_ready()
    t6 = %timeit -o jit_gradv(X, U, V).block_until_ready()
    time_jit[i, 0] = t5.average
    time_jit[i, 1] = t6.average
    #print(t, i, time.time() - start_time)
````

_2._ (2 pts) Derive analytical expression for the gradient of the function $f$:

$$ R(x) = \frac{(Ax, x)}{(x, x)}, $$

where $A$ is a symmetric real matrix. Why the gradient of this function is important in NLA you will know in the lectures later.

#### Solution
$\dfrac{dR(x)}{dx} = \dfrac{\dfrac{d(Ax, x)}{dx} \cdot (x, x) - (Ax, x) \cdot \dfrac{d(x, x)}{dx}}{(x, x)^2}$

$\dfrac{d(x, x)}{dx} = \dfrac{d(\sum\limits_{i=1}^{n}{x_i^2})}{dx} = \Big({\dfrac{d(x_1^2)}{dx_1}, \dfrac{d(x_2^2)}{dx_2}, ... \dfrac{d(x_n^2)}{dx_n}}\Big)^T = 2 \cdot x$

$\dfrac{d(Ax, x)}{dx} = \dfrac{d(x^TAx)}{dx} = (*)$

$x^TAx = (x_1, x_2,... x_n) \cdot \begin{bmatrix} a_{11}& a_{12} & ... & a_{1n}\\ a_{21}& a_{22} & ... & a_{2n}\\ ... & ... & ... & ...\\ a_{n1}& a_{n2} & ... & a_{nn} \end{bmatrix} \cdot \begin{pmatrix} x_1\\x_2\\...\\x_n \end{pmatrix} = \Big({\sum\limits_{k=1}^{n}{a_{k1}x_k}, \sum\limits_{k=1}^{n}{a_{k2}x_k}}, ..., \sum\limits_{k=1}^{n}{a_{kn}x_k}\Big) \cdot \begin{pmatrix} x_1\\x_2\\...\\x_n \end{pmatrix} = \sum\limits_{k=1}^{n} \sum\limits_{m=1}^{n} a_{km} \cdot x_k \cdot x_m$

$\dfrac{d(x^TAx)}{dx_l} = \sum\limits_{j=1}^{n}{x_j(a_{jl} + a_{lj})}$

$(*) = (A + A^T) x$

$\dfrac{dR(x)}{dx} = \dfrac{(x, x) \cdot (A + A^T) x - 2 \cdot (Ax, x) x}{(x, x) ^ 2}$

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.fc2(x)
        x = self.ReLU(x)
        x = self.fc3(x)
        x = self.ReLU(x)
        x = self.fc4(x)
        x = self.ReLU(x)
        x = self.fc5(x)
        x = self.ReLU(x)
        x = self.fc6(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#           data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

````python
class compressed_net(nn.Module):
    def __init__(self, model, r):
        super(compressed_net, self).__init__()
        self.layers = []
        for layer in list(model.modules()):
            if (type(layer) == nn.modules.linear.Linear):
                A = layer.weight.data
                bias = layer.bias.data
                u, s, v = torch.svd(A)
                V = nn.Linear(v.shape[0], r, bias = False)
                V.weight.data = v.t()[:r]
                S = nn.Linear(r, r, bias = False)
                S.weight.data = torch.diag(s[:r])
                U = nn.Linear(r, u.shape[0]) # bias = False)
                U.weight.data = u.t()[:r].t()
                U.bias.data = layer.bias.data
                self.layers.append((V, S, U))
                
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = x.view(-1, 3 * 32*32)
        
        for i in range(len(self.layers)):
            for j in range(3):
                x = self.layers[i][j](x)
            if (i < len(self.layers) - 1):
                x = self.ReLU(x)
            
        return F.log_softmax(x, dim=1)
````

````python
import torch
import jax.numpy as jnp
import jax
import numpy as np
import random
#from collections import Counter

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)
    def hermione_computes_function(self, x):
        return jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))
    def harry_answers(self):
        m = 10
        result = 0
        for i in range(m):
            key, subkey = jax.random.split(self.key)
            x_ = jax.random.uniform(subkey, (100, 1), maxval=10., minval=0., dtype=jnp.float32)
            grad_ = jax.jacfwd(self.hermione_computes_function)(x_)
            grad_ = np.array(np.reshape(grad_ / np.sqrt(m), (100, 100))) 
            G = torch.from_numpy(grad_)
            result += jax.numpy.linalg.matrix_rank(G).item()
        r = int(result / m)
        return r
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````

#### Check that it's a norm by a definition:
$1. \| x \| = 0 \Leftrightarrow \forall i \to x_i - x_{i-1} = 0 \Leftrightarrow \forall i \to x_i = 0 \Leftrightarrow x = 0 $
